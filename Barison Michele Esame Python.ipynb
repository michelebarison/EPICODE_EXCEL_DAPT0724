## Esercizio 2
Si chiede di trovare, per ogni continente: 

a. il numero di casi fin dall'inizio della pandemia

b. la percentuale rispetto al totale mondiale del numero di casi

```python
# Rimuoviamo eventuali valori NaN nella colonna "continent"
covid_clean = covid.dropna(subset=["continent"])

# Calcoliamo il numero totale di casi per continente utilizzando groupby (avrei potuto utilizzare anche una pivot)
cases_per_continent = covid_clean.groupby("continent")["total_cases"].max()

# Calcoliamo il totale mondiale dei casi
total_world_cases = cases_per_continent.sum()

# Calcoliamo la percentuale di ogni continente rispetto al totale mondiale
cases_percentage = (cases_per_continent / total_world_cases) * 100

# Stampiamo i risultati
print("Numero di casi per continente:\n", cases_per_continent)
print("\nPercentuale rispetto al totale mondiale:\n", cases_percentage)
```

## Esercizio 3

Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati:

a. l'evoluzione del casi totali dall'inizio alla fine dell'anno 

b. il numero di nuovi casi rispetto alla data; poiché i nuovi casi vengono registrati settimanalmente, filtrare via i dati che non hanno misurazioni

```python
import matplotlib.pyplot as plt


# Filtriamo i dati per l'Italia nel 2022
italy_2022 = covid.query("location == 'Italy' and '2022-01-01' <= date <= '2022-12-31'").copy()

# Convertiamo la colonna "date" in formato data
italy_2022["date"] = pd.to_datetime(italy_2022["date"])

# Grafico casi totali
plt.plot(italy_2022["date"], italy_2022["total_cases"])
plt.xlabel("Data")
plt.ylabel("Casi Totali")
plt.title("Evoluzione dei casi totali in Italia (2022)")
plt.xticks(rotation=45)
plt.show()

# Grafico nuovi casi (senza NaN)
plt.plot(italy_2022.dropna(subset=["new_cases"])["date"], italy_2022.dropna(subset=["new_cases"])["new_cases"])
plt.xlabel("Data")
plt.ylabel("Nuovi Casi")
plt.title("Nuovi casi giornalieri in Italia (2022)")
plt.xticks(rotation=45)
plt.show()
```

## Fine degli Esercizi

Andiamo oltre...
In Europa, quali sono stati gli Stati maggiormente colpiti dall'epidemia? Ecco la TOP 10

```python
# Filtriamo solo gli stati europei (continent = 'Europe') ed eliminiamo i valori nulli
european_covid = covid.dropna(subset=["continent", "total_cases"])
european_covid = european_covid[european_covid["continent"] == "Europe"]

# Creiamo una pivot table per ottenere il massimo numero di contagi per stato
europe_cases_pivot = european_covid.pivot_table(values="total_cases", index="location", aggfunc="max")

# Selezioniamo i 10 stati europei con il numero più alto di contagiati
top_10_european_states = europe_cases_pivot.nlargest(10, "total_cases")

# Creiamo il grafico a barre
plt.figure(figsize=(12, 6))
plt.bar(top_10_european_states.index, top_10_european_states["total_cases"], color='blue')
plt.xlabel("Stato Europeo")
plt.ylabel("Numero totale di contagiati")
plt.title("Top 10 Stati Europei con il numero più alto di contagiati")
plt.xticks(rotation=45, ha='right')
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostriamo il grafico
plt.show()
```

# Andiamo a vederli per anno - 2020 - 2021 - 2022 - 2023 

```python
# Filtriamo solo gli stati europei e rimuoviamo i valori nulli
european_covid = covid.dropna(subset=["continent", "total_cases"])
european_covid = european_covid[european_covid["continent"] == "Europe"]

# Convertiamo la colonna "date" in formato datetime
european_covid["date"] = pd.to_datetime(european_covid["date"])

# Creiamo una funzione per generare il grafico per un dato anno
def plot_top_european_countries(year):
    # Filtriamo i dati per l'anno selezionato
    yearly_data = european_covid[european_covid["date"].dt.year == year]
    
    # Creiamo una pivot table per ottenere il massimo numero di contagi per stato
    cases_pivot = yearly_data.pivot_table(values="total_cases", index="location", aggfunc="max")
    
    # Selezioniamo i 10 stati con il numero più alto di contagiati
    top_10_states = cases_pivot.nlargest(10, "total_cases")

    # Creiamo il grafico a barre
    plt.figure(figsize=(12, 6))
    plt.bar(top_10_states.index, top_10_states["total_cases"], color='red')
    plt.xlabel("Stato Europeo")
    plt.ylabel("Numero totale di contagiati")
    plt.title(f"Top 10 Stati Europei con il numero più alto di contagiati ({year})")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Generiamo i grafici per gli anni 2020, 2021 e 2022
for year in [2020, 2021, 2022]:
    plot_top_european_countries(year)
```
